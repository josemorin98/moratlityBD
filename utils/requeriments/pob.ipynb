```python
import pandas as pd
import numpy as np
```

```python
df_pob = pd.read_csv('./poblaciones_edad_state.csv')
df_pob.columns
```

```python
df_pob.loc[df_pob['ANIO_REGIS']==2015]
```

```python
colsR = ['TOTAL', '00_04', '05_14', '15_24', '25_44', '45_64', '>65']
for columnsPob in colsR:
    df_pob[columnsPob] = np.where(df_pob['ANIO_REGIS'] == 2015, np.round(df_pob[columnsPob]/2,0), df_pob[columnsPob])
    df_pob[columnsPob] = df_pob[columnsPob].astype(int)
df_pob.loc[df_pob['ANIO_REGIS']==2015]
```

```python
df_pob.to_csv('./poblaciones_edad_state_new.csv',index=False)
```
